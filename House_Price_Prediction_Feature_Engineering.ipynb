#MISSING VALUES

#####CATEGORICAL MISSING VALUES:

```python
#finding out the features which have nan values
cat_nan_features=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtype == "O"]
cat_nan_features
```

```python
for feature in cat_nan_features:
  print("missing values % in {} is : {}%".format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
#replacing missing values with new label:
def replace_cat_features(data,cat_nan_features):
  data1=data.copy()
  data1[cat_nan_features] = data1[cat_nan_features].fillna("missing")
  return data1
data = replace_cat_features(data,cat_nan_features)
data[cat_nan_features].isnull().sum()
```

```python
 data.head()
```

#####NUMERICL MISSING VALUES:

```python
#checking for missing  numerical values:
num_cat_features = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtype !="O"]
num_cat_features
```

```python
for feature in num_cat_features:
  print("% of missing values in {} is : {}%".format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
#replacing missing values in numerical features with nan:
#since we have outliers in continious variables replacing nan with median or mean:
for feature in num_cat_features:
  #finding the median value in each feature
  median_value =  data[feature].median()
  print('median value for feature {} is : {}'.format(feature,median_value))
  data[feature+'nan'] = np.where(data[feature].isnull(),1,0)  #creating new feature to show if nan value if present as 1 or if not as 0
  data[feature].fillna(median_value,inplace=True) #replcing na with missing value
data[num_cat_features].isnull().sum()
```

```python
data.head(10)
```

#####handling temporal data features

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  data[feature]= data["YrSold"]-data[feature]
```

```python
data.head()
```

#####Transforming numerical features which are skwed

```python
#since numerical features are skewed transforming those features with some transformation techniques:
num_features = ["LotFrontage","LotArea","1stFlrSF","GrLivArea","SalePrice"]
for feature in num_features:
  data[feature]=np.log(data[feature])
```

```python
data.head()
```

#####Handling rare categorical features

```python
categorical_features = [feature for feature in data.columns if data[feature].dtype=="O"]
categorical_features
```

```python
for feature in categorical_features:
  temp = data.groupby(feature)["SalePrice"].count()/len(data)
  temp_df = temp[temp>0.01].index
  data[feature] = np.where(data[feature].isin(temp_df),data[feature],"Rare_feature")
```

```python
data.head(50)
```

```python
#converting categorical varaibles to numerical varaibles:
for feature in categorical_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head(50)
```

#####FEATURE SCALING

```python
scaling_features = [feature for feature in data.columns if feature not in ["Id","SalePrice"]]
print("no of features need to be scaled : {}".format(len(scaling_features)))
scaling_features
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[scaling_features])
```

```python
#concading the scaled features and id and scaleprice features to x_train 
x_train = pd.concat([data[["Id","SalePrice"]],pd.DataFrame(scaler.transform(data[scaling_features]),columns = scaling_features)],axis=1)
```
